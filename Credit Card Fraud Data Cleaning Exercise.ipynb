```python
corr = credit_df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
corr_df = credit_df.drop("Class", axis=1).corr().abs()
pairwise_corr= (corr_df.where(np.triu(np.ones(corr_df.shape), k=1).astype(np.bool))
                 .stack()
                 .sort_values(ascending=False))
pairwise_corr
```

```python
## Exploring class/target variable for class imbalance
catVar = 'Class'
fig, axes12 = plt.subplots(1,2,figsize=(15,6))
fig.suptitle('Distribution of Class / Target Variable Among Its Different Levels', fontsize=20)
credit_df[catVar].value_counts().plot(kind='bar',ax=axes12[0])
credit_df[catVar].value_counts(normalize=True).plot(kind='barh', ax=axes12[1])
fig, axes12 = plt.subplots(1,1,figsize=(15,6))
credit_df[catVar].value_counts().plot(kind='pie',autopct='%1.1f%%', shadow=True, startangle=90)
axes12.axis('equal')
plt.show()
```
